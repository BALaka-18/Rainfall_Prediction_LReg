```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
missing_values = [0,'NA']
dataset = pd.read_csv("C:\\Users\\BIPLAB\\Desktop\\rainfall in india 1901-2015.csv", na_values = missing_values)
dataset                                   
```

```python
print("No. of null values: ", dataset.isnull().sum())
print("Filling null values with mean of that column:")
dataset = dataset.fillna(np.mean(dataset))
dataset
```

```python
dataset.groupby('SUBDIVISION').size()
```

```python
print("Co-variance = ", dataset.cov())
print("\nCo-relation = ", dataset.corr())
```

```python
print("Comparative plotting between Annual Rainfall in all states and that in January in all states.")
plt.scatter(dataset.ANNUAL, dataset.JAN)
```

```python
print("Informative plotting of the Annual Rainfall in all states.")
dataset['ANNUAL'].hist(bins = 20)
```

```python
from sklearn import linear_model, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
y = dataset['ANNUAL']
x = dataset[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']]
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size = 0.3, shuffle = False)
print("Train_x shape", train_x.shape, "\nTest_x shape", test_x.shape)
print("Train_y shape", train_y.shape, "\nTest_y shape", test_y.shape)
```

```python
lm = linear_model.LinearRegression()
lm.fit(train_x,train_y)
predicted = lm.predict(test_x)
print("Actual : ", test_y)
print("\nPredicted : ", predicted)
```

```python
r2scr = r2_score(test_y, predicted)
print("R2_score = ", r2scr)
rmse = np.sqrt(mean_squared_error(test_y, predicted))
print("\nRoot mean squared error = ", rmse)
```
